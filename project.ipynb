```python
passingmark=33
sp['pass_math'] = np.where(sp['math score']>= passingmark, 'Pass', 'Fail')
sp['pass_math'].value_counts(dropna = False).plot.bar(color=['green','red'], figsize = (3,3))

plt.title('Comparison between Pass/Fail in Maths')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
sp['pass_math'].value_counts()
```

```python
sp['pass_read'] = np.where(sp['reading score']>= passingmark, 'Pass', 'Fail')
sp['pass_read'].value_counts(dropna = False).plot.bar(color = ['green','red'], figsize = (3,3))
plt.title('Comparison between Pass/Fail in Reading')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
sp['pass_read'].value_counts()
```

```python
sp['pass_write'] = np.where(sp['writing score']>= passingmark, 'Pass', 'Fail')
sp['pass_write'].value_counts(dropna = False).plot.bar(color = ['green','red'], figsize = (3,3))
plt.title('Comparison between Pass/Fail in Writing')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
sp['pass_write'].value_counts()
```

```python
sp['total_score'] = sp['math score'] + sp['reading score'] + sp['writing score']

sp['total_score'].value_counts(normalize = True)
sp['total_score'].value_counts(dropna = True).plot.bar(color = 'green', figsize = (40,10))

plt.title('Comparison Of Total Score Of All Students')
plt.xlabel('Total Score')
plt.ylabel('Count')
plt.show()
```

```python
# Sample DataFrame creation (you should load your actual data instead)
# Example: sp = pd.read_csv('your_data.csv')
sp = pd.DataFrame({
    'student_id': [1, 2, 3],
    'total_score': [210, 180, 240]  # example total out of 300
})

# Calculate percentage
sp['percentage'] = sp['total_score'] / 3
sp['percentage'] = sp['percentage'].apply(ceil)

print(sp)
```

```python
#Assigning Pass/Fail through overall score

sp['result'] = sp.apply(lambda x : 'Fail' if x['pass_math'] == 'Fail' or 
                           x['pass_read'] == 'Fail' or x['pass_write'] == 'Fail'
                           else 'Pass', axis = 1)

sp['result'].value_counts(dropna = False).plot.bar(color = ['green','red'], figsize = (3, 3))
plt.title('Overall Result')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
sp['result'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (20,10)

# First subplot
plt.subplot(1, 3, 1)
sns.swarmplot(x='result', y='total_score', hue='gender', data=sp, palette='bright')
plt.title('Result vs Total Score', fontsize=20)

# Second subplot
plt.subplot(1, 3, 2)
sns.swarmplot(x='lunch', y='total_score', hue='gender', data=sp, palette='bright')
plt.title('Type of Lunch vs Total Score', fontsize=20)

# Third subplot
plt.subplot(1, 3, 3)
sns.swarmplot(x='test preparation course', y='total_score', hue='gender', data=sp, palette='bright')
plt.title('Test Preparation Course vs Total Score', fontsize=20)

# Show all plots
plt.tight_layout()
plt.show()
```
